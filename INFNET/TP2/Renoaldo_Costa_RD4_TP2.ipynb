````python
# Importa o m√≥dulo para trabalhar com vari√°veis de ambiente
# !pip install python-dotenv
import os

# Importa a biblioteca Google Generative AI para interagir com o modelo LLM da Google
import google.generativeai as genai

# Importa o m√≥dulo "userdata" do Google Colab para acessar dados do usu√°rio, como a chave da API
from google.colab import userdata

# Configura a biblioteca GenAI com a chave da API armazenada na vari√°vel de ambiente 'GEMINI_KEY'
genai.configure(api_key=userdata.get('GEMINI_KEY'))

# Especifica o modelo generativo que ser√° usado, neste caso, "gemini-1.5-flash"
model = genai.GenerativeModel("gemini-1.5-flash")

# Texto da not√≠cia que ser√° resumido
noticia = """
Aprovadas contas anuais de mais duas prefeituras
7 de novembro de 2024
Na sess√£o desta quinta-feira (07/11), os conselheiros do Tribunal de Contas dos Munic√≠pios da Bahia recomendaram √†s c√¢maras de vereadores a aprova√ß√£o com ressalvas de mais duas presta√ß√µes de contas anuais de prefeituras ‚Äì uma relativa ao exerc√≠cio de 2023 e a outra de 2020.
As contas aprovadas s√£o da prefeitura de Mutu√≠pe, de 2020, de responsabilidade de Rodrigo Maicon de Santana Andrade ‚Äì que foi multado em R$ 2 mil; e de Dom Bas√≠lio, de 2023, do prefeito Roberval de C√°ssia Meira.
Foram avaliadas as contas de governo e gest√£o dessas prefeituras, sendo apontadas algumas ressalvas que n√£o interferiram, no entanto, no m√©rito do julgamento pela aprova√ß√£o.
Cabe recurso das decis√µes.
"""

# Define o prompt que ser√° passado ao modelo. Inclui a not√≠cia e pede um resumo em at√© 2 frases.
# O prompt segue um exemplo de estrutura para o modelo entender o contexto e a forma desejada do resumo.
prompt = f"""
Resuma a <NOTICIA> em at√© 2 senten√ßas com os principais pontos da noticia.
Aqui est√° um exemplo de resumo:

<NOTICIA>
"A empresa X lan√ßou um novo smartphone no mercado."
###
<RESUMO>
"A empresa X anunciou o lan√ßamento de um novo smartphone, trazendo inova√ß√µes tecnol√≥gicas e maior dura√ß√£o de bateria."

Agora, fa√ßa o resumo da noticia abaixo:
<NOTICIA>
"{noticia}"
###
<RESUMO>
"""

# Gera o resumo da not√≠cia usando o modelo configurado
response = model.generate_content(prompt)

# Extrai o texto do resumo gerado pelo modelo
text_response = response.to_dict()["candidates"][0]["content"]["parts"][0]["text"]

# Exibe o resumo gerado pelo modelo LLM
print("Resumo gerado pelo LLM:")
print(text_response)
````

````python
import yaml
from google.colab import userdata
import google.generativeai as genai
import json
import pandas as pd
import plotly.express as px

# Obter a chave da API a partir das vari√°veis de ambiente
api_key = userdata.get('GEMINI_KEY')

if not api_key:
    raise ValueError("A chave da API 'GEMINI_KEY' n√£o foi encontrada nas vari√°veis de ambiente.")

# Configura a biblioteca GenAI para utilizar a chave da API
genai.configure(api_key=api_key)

# Define o modelo de linguagem generativa que ser√° usado
model = genai.GenerativeModel("models/gemini-1.5-flash")

# Fun√ß√£o para executar o prompt usando o modelo generativo configurado
def execute_prompt(city_prompt):
    response = model.generate_content(city_prompt)
    return response.text.strip()

# Carregar o arquivo YAML
with open('roteiro_turismo_json.yaml', 'r') as file:
    prompts = yaml.safe_load(file)

# Executar os prompts para Atenas e Roma
results = {}
for city, prompt_text in prompts['roteiro'].items():
    print(f"Executando prompt para {city.capitalize()}...")
    results[city] = execute_prompt(prompt_text)

# Exibir os resultados brutos
for city, response in results.items():
    print(f"\n--- Resultados para {city.capitalize()} ---")
    print(response)

# Fun√ß√£o para processar cada string JSON e carregar em um DataFrame
def process_results_to_df(results):
    dfs = []
    for city, json_string in results.items():
        # Remover as marca√ß√µes de c√≥digo
        json_string_clean = json_string.replace("```json\n", "").replace("\n```", "").strip()

        # Verificar se a string JSON est√° vazia
        if not json_string_clean:
            print(f"A resposta para {city.capitalize()} est√° vazia ap√≥s limpar as marca√ß√µes.")
            continue

        # Carregar a string JSON em um objeto Python (lista de dicion√°rios)
        try:
            data = json.loads(json_string_clean)
        except json.JSONDecodeError as e:
            print(f"Erro ao decodificar JSON para a cidade {city.capitalize()}: {e}")
            continue

        # Verificar se os dados s√£o uma lista
        if not isinstance(data, list):
            print(f"Dados para {city.capitalize()} n√£o s√£o uma lista. Tipo encontrado: {type(data)}")
            continue

        # Converter os dados para um DataFrame
        df = pd.DataFrame(data)

        # Adicionar uma coluna 'cidade'
        df['cidade'] = city.capitalize()

        # Adicionar o DataFrame √† lista
        dfs.append(df)

    # Verificar se h√° DataFrames para concatenar
    if not dfs:
        print("Nenhum DataFrame foi criado a partir dos resultados.")
        return pd.DataFrame()  # Retorna um DataFrame vazio

    # Concatenar todos os DataFrames em um √∫nico DataFrame final
    final_df = pd.concat(dfs, ignore_index=True)

    return final_df

# Criar o DataFrame final
final_df = process_results_to_df(results)

# Exibir o DataFrame final
print("\n--- DataFrame Final ---")
display(final_df)

# Criando o gr√°fico de barras horizontais
fig = px.bar(final_df, x='visitantes_anuais', y='ponto_turistico', color='cidade', orientation='h')

fig.update_layout(
    title="Visitantes Anuais por Cidade e Ponto Tur√≠stico",
    xaxis_title="Visitantes Anuais",
    yaxis_title="Ponto Tur√≠stico"
)

# Exibindo o gr√°fico
fig.show()
````
